```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt 
from PIL import Image
from scipy import misc
import scipy.ndimage as ndimage
import cv2 as cv
#cv.subtract resta las imagenes
```

```python
#Importamos la imagen del tigre
tigre = Image.open(r'/home/nathan/VisionporComputador/lab4/Tigre.jpg')
#Importamos la imagen del lobo
lobo = Image.open(r'/home/nathan/VisionporComputador/lab4/Lobo.jpg')
#Veamos el tamaño de las imagenes
print('Resolución tigre = '+str(np.asarray(tigre).shape[0])+" x "+str(np.asarray(tigre).shape[1]))
print('Resolución lobo = '+str(np.asarray(lobo).shape[0])+" x "+str(np.asarray(lobo).shape[1]))

#Las dejamos de las mismas dimensiones 
ancho = np.min([np.asarray(tigre).shape[0],np.asarray(lobo).shape[0]])
largo = np.min([np.asarray(tigre).shape[1],np.asarray(lobo).shape[1]])
tigre=tigre.resize((largo,ancho))
lobo=lobo.resize((largo,ancho))
#Suavizamos el tigre
tigre=ndimage.gaussian_filter(tigre, sigma=(7, 7, 0), order=0)
#Agarramos solo las frecuencias altas del lobo
lobo=cv.subtract(np.asarray(lobo),np.asarray(ndimage.gaussian_filter(lobo, sigma=(2, 2, 0), order=0)))
```

```python
tigreA=np.asarray(tigre)
loboA=np.asarray(lobo)
```

```python
plt.figure()
plt.imshow(tigreA,cmap='gray')
plt.axis('off')
plt.savefig("Tigresuave.jpg")
plt.show()
plt.close
plt.figure()
plt.imshow(loboA,cmap='gray')
plt.axis('off')
plt.savefig("Loboalt.jpg")
plt.show()
plt.close()
```

```python
#Ahora sumamos las imagenes
Combinada=cv.add(tigreA,loboA)
plt.imshow(Combinada)
plt.show()
```

```python
plt.imshow(loboA,cmap='gray')
plt.show()
#Ahora reflejemos la imagen respecto al centro
lobomirror = Image.fromarray(loboA).transpose(Image.FLIP_LEFT_RIGHT)
loboA= np.asarray(lobomirror)
plt.imshow(loboA,cmap='gray')
plt.show()
#Quiero correr el lobo a la izquierda así que voy a usar un filtro para hacerlos
correrizquierda=np.array([[0,0,0],[0,0,1],[0,0,0]])
#Ahora calculo la cross correlación entre el lobo y mi filtro
for i in range(5):
    loboA=cv.filter2D(loboA,-1,correrizquierda)
plt.imshow(loboA,cmap='gray')
plt.show()

```

```python
#Volvemos a combinar 
#Ahora sumamos las imagenes
Combinada=cv.add(tigreA,loboA)
plt.imshow(Combinada)
plt.show()
```

```python
#Volvemos Combinada una imagen y la recortamos
CombinadaI=Image.fromarray(Combinada)
CombinadaI=CombinadaI.crop((0,0,225,177))
Combinada=np.asarray(CombinadaI)
plt.figure()
plt.imshow(Combinada)
plt.axis('off')
plt.savefig("Lobotigre.jpg")
plt.show()
```

```python
#Hacemos la piramide Gaussiana
G = Combinada.copy()
gpcombinada = [G]
for i in range(6):
    G = cv.pyrDown(G)
    gpcombinada.append(G)
cont=0
for i in gpcombinada:
    plt.figure()
    plt.imshow(i,cmap="gray")
    plt.axis('off')
    plt.savefig("TigreloboGa"+str(cont)+".jpg")
    plt.show()
    plt.close()
    cont+=1
```

```python

#Ahora hacemos el blending de dos imagenes 
#Importamos la imagen de Batman
batman = Image.open(r'/home/nathan/VisionporComputador/lab4/batman.jpg')
batman=batman.crop((325,70,398,220))
#Importamos la imagen mia
yo = Image.open(r'/home/nathan/VisionporComputador/lab4/yo.jpg')
yo= rotate(yo,90)
yo= rotate(yo,90)
yo=Image.fromarray(yo).crop((900,150,1650,1550))
yo=yo.resize((96,160))
batman=batman.resize((96,160))
#Veamos el tamaño de las imagenes
print('Resolución batman = '+str(np.asarray(batman).shape[0])+" x "+str(np.asarray(batman).shape[1]))
print('Resolución yo = '+str(np.asarray(yo).shape[0])+" x "+str(np.asarray(yo).shape[1]))
#Las volvemos arrays
batmana=np.asarray(batman)
yoa=np.asarray(yo)
```

```python
plt.imshow(batmana,cmap='gray')
plt.show()
plt.imshow(yoa,cmap='gray')
plt.show()
```

```python
#Generamos las piramides Gaussianas
Batman=batmana.copy()
gpb= [Batman]
for i in range(3):
    Batman = cv.pyrDown(Batman)
    gpb.append(Batman)
Yo=yoa.copy()
gpy= [Yo]
for i in range(3):
    Yo = cv.pyrDown(Yo)
    gpy.append(Yo)
cont=0
for i in gpb:
    plt.figure()
    plt.imshow(i,cmap="gray")
    plt.axis('off')
    plt.savefig("BatmanGa"+str(cont)+".jpg")
    plt.show()
    plt.close()
    cont+=1
cont=0
for i in gpy:
    plt.figure()
    plt.imshow(i,cmap="gray")
    plt.axis('off')
    plt.savefig("YoGa"+str(cont)+".jpg")
    plt.show()
    plt.close()
    cont+=1
```

```python
#Generamos la piramide Laplaciana
lpb=[gpb[3]]
lista=[3,2,1]
for i in lista:
    GR= cv.pyrUp(gpb[i])
    La=cv.subtract(np.asarray(gpb[i-1]),np.asarray(GR))
    lpb.append(La)
cont=0
for i in lpb:
    plt.figure()
    plt.imshow(i,cmap="gray")
    plt.axis('off')
    plt.savefig("BatmanLa"+str(cont)+".jpg")
    plt.show()
    plt.close()
    cont+=1
    
lpy=[gpy[3]]
for i in lista:
    GR= cv.pyrUp(gpy[i])
    La=cv.subtract(np.asarray(gpy[i-1]),np.asarray(GR))
    lpy.append(La)
cont=0
for i in lpy:
    plt.figure()
    plt.imshow(i,cmap="gray")
    plt.axis('off')
    plt.savefig("YoLa"+str(cont)+".jpg")
    plt.show()
    plt.close()
    cont+=1
```

```python
#Código tomado de https://docs.opencv.org/3.1.0/dc/dff/tutorial_py_pyramids.html
LS = []
for la,lb in zip(lpb,lpy):
    rows,cols,dpt = la.shape
    ls = np.hstack((la[:,0:cols/2], lb[:,cols/2:]))
    LS.append(ls)
cont=0
for i in LS:
    plt.figure()
    plt.imshow(i,cmap="gray")
    plt.axis('off')
    plt.savefig("Unidas"+str(cont)+".jpg")
    plt.show()
    plt.close()
    cont+=1
    
```

```python
#Ahora empezmos a subir la gaussiana junta y le sumamos la laplaciana correspondiente. Este pedazo también fue inspiracion de 
# https://docs.opencv.org/3.1.0/dc/dff/tutorial_py_pyramids.html
ls_ = LS[0]
for i in [1,2,3]:
    ls_ = cv.pyrUp(ls_)
    ls_ = cv.add(ls_, np.asarray(Image.fromarray(LS[i]).resize((ls_.shape[1],ls_.shape[0]))))
plt.figure()
plt.imshow(ls_,cmap='gray')
plt.axis('off')
plt.savefig('yoba.jpg')
plt.show()
```
