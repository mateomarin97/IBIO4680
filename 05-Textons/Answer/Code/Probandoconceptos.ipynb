```python
#Esta funcion pasa la matriz de textones a un histograma
def histc(X, bins):
    map_to_bins = np.digitize(X,bins)
    r = np.zeros(bins.shape)
    for i in map_to_bins:
        r[i-1] += 1
    return np.array(r)
```

```python
#Hagamos los histogramas
histladrillo= histc(tmapladrillo.flatten(),np.arange(k))
print(histladrillo)
#Normalizamos
histladrillo= histc(tmapladrillo.flatten(),np.arange(k))/tmapladrillo.size
print(histladrillo)
```

```python
#Definimos el kernel de la interseccion
def kernelint(A,B):
    if(len(A)!=len(B)):
        print("Error, los dos histogramas deben tener el mismo tamaño")
    else:
        suma=0
        for i in range(len(A)):
            suma+=min(A[i],B[i])
        return suma
            
```

```python
#Hagamos los histogramas
histladrillo4= histc(tmapladrillo4.flatten(),np.arange(k))
print(histladrillo4)
#Normalizamos
histladrillo4= histc(tmapladrillo4.flatten(),np.arange(k))/tmapladrillo4.size
print(histladrillo4)
#Definimos el coeficiente de similitud entre los dos histogramas
coef1=kernelint(histladrillo,histladrillo4)
print(coef1)
```
